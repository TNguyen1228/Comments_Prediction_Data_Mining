```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.tree import DecisionTreeRegressor
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#data train
file_train='D:\\Project\\Data_Mining\\Finals_BlogCommentPrediction\\Data_Ucimlrepo\\Train\\blogData_train.csv'
train=pd.read_csv(file_train, header=None)

#data test
file_test='D:\\Project\\Data_Mining\\Finals_BlogCommentPrediction\\Data_Ucimlrepo\\Test\\blogData_test.csv'
test=pd.read_csv(file_test, header=None)

train = train.drop(train.columns[:50], axis=1).drop(train.columns[55:60], axis=1).drop(train.columns[277:280], axis=1)  # Drop the first 50 columns
test = test.drop(test.columns[:50], axis=1).drop(test.columns[55:60], axis=1).drop(test.columns[277:280], axis=1)  # Drop the first 50 columns

print(train.isnull().values.any())
print(test.isnull().values.any())
test.fillna(value=0, inplace=True)
print(test.isnull().values.any())
# Split features and labels
X_train = train.iloc[:, :-1]
y_train = train.iloc[:, -1]
X_test = test.iloc[:, :-1]
y_test = test.iloc[:, -1]
X_test
```

```python
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred
```

```python
import joblib

joblib.dump(model, 'decision_tree_regressor.joblib')
```

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error
# Calculate MSE
mse = mean_squared_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)  # Root Mean Squared Error


print("Decision tree regression result")
print("Mean Squared Error:", mse)
print(f'RMSE : {rmse}')

mae = mean_absolute_error(y_test, y_pred)

print(f'Mean Absolute Error (MAE): {mae}')
```
